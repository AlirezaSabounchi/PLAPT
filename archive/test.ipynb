```python
import os

# Specify the directory you want to list
directory = "C:\\Users\\tatwo\\Downloads\\Source Code of \
CAPLA\\CAPLA\\data\\CSAR-HiQ_36\\global"

# List all files and directories in the specified directory
files_and_directories = os.listdir(directory)

# Get full file paths
protein_csvs = [os.path.join(directory, f) for f in files_and_directories if os.path.isfile(os.path.join(directory, f))]
```

```python
import pandas as pd

# Function to extract protein sequence from a CSV file
def extract_protein_sequence(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path)

    # Select columns that start with 'a_'
    protein_data = df.filter(regex='^a_')

    # Process the extracted data as needed
    # For example, return the dataframe
    return protein_data


# Function to map columns to amino acids
def column_to_amino_acid(column):
    amino_acid_map = {
        'a_G': 'G', 'a_A': 'A', 'a_V': 'V', 'a_L': 'L', 'a_I': 'I',
        'a_M': 'M', 'a_F': 'F', 'a_P': 'P', 'a_W': 'W', 'a_S': 'S',
        'a_T': 'T', 'a_Y': 'Y', 'a_C': 'C', 'a_Q': 'Q', 'a_N': 'N',
        'a_D': 'D', 'a_E': 'E', 'a_K': 'K', 'a_R': 'R', 'a_H': 'H',
        'a_X': 'X'
    }
    return amino_acid_map.get(column, 'X')

# Function to convert 1-hot encoded DataFrame to amino acid string
def one_hot_to_amino_acid_string(protein_sequence):
    amino_acid_string = ''

    for index, row in protein_sequence.iterrows():
        # Find the column where the value is 1
        for column in protein_sequence.columns:
            if row[column] == 1:
                amino_acid_string += column_to_amino_acid(column)
                break

    return amino_acid_string

prot_sequences={}
# Loop through each file and process
for file_path in protein_csvs:
    protein_sequence = extract_protein_sequence(file_path)
    amino_acid_string = one_hot_to_amino_acid_string(protein_sequence)
    prot_sequences[file_path[-8:-4]] = amino_acid_string
```

```python
import json
with open('prot_sequences.json', 'w') as file:
    json.dump(prot_sequences, file)
```
